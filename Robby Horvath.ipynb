```python
import pandas as pd
%matplotlib inline

arrest_data = pd.read_csv("Non_Traffic_Citations.csv",
                                       index_col="CITEDTIME", parse_dates=True)

arrest_neighborhoods = arrest_data['NEIGHBORHOOD'].value_counts()
arrest_df = pd.DataFrame(arrest_neighborhoods)
arrest_hood = arrest_df.rename(columns={"NEIGHBORHOOD": "Citation Count Since 2016"})
arrest_hood.index.name = "Neighborhood"
arrest_hood.head()
```
