```python
import pandas as pd

citation_data = pd.read_csv("Non_Traffic_Citations.csv",
                                       index_col="CITEDTIME", parse_dates=True)

citation_neighborhoods = citation_data['NEIGHBORHOOD'].value_counts()
citation_df = pd.DataFrame(citation_neighborhoods)
citation_hood = citation_df.rename(columns={"NEIGHBORHOOD": "Citation Count Since 2014"})
citation_hood.index.name = "Neighborhood"
citation_hood.head()
```

```python
plot = citation_neighborhoods.plot(kind="bar", width = 0.4, figsize=(25,15), title="Citations per Neighborhood in Allegheny County since 2014")
plot.set_ylabel("# of Citations")
plot.set_xlabel("Neighborhood")
```
